**(b, 8 points)** What is the range, mean and standard deviation of each quantitative
predictor? Which month has the highest number of fires?

```python
import pandas as pd
forest_fires = pd.read_csv('forestfires.csv')
# print(forest_fires)

# FFMC
ffmc_range = forest_fires['FFMC'].max() - forest_fires['FFMC'].min()
ffmc_mean = forest_fires['FFMC'].mean()
ffmc_standard_deviation = forest_fires['FFMC'].std()

print(f"FFMC range: {ffmc_range}")
print(f"FFMC mean: {ffmc_mean}")
print(f"FFMC standard deviation: {ffmc_standard_deviation}")

# DMC
dmc_range = forest_fires['DMC'].max() - forest_fires['DMC'].min()
dmc_mean = forest_fires['DMC'].mean()
dmc_standard_deviation = forest_fires['DMC'].std()

print(f"DMC range: {dmc_range}")
print(f"DMC mean: {dmc_mean}")
print(f"DMC standard deviation: {dmc_standard_deviation}")

# DC
forest_fires['DC'] = pd.to_numeric(forest_fires['DC'], errors='coerce')
dc_range = forest_fires['DC'].max() - forest_fires['DC'].min()
dc_mean = forest_fires['DC'].mean()
dc_standard_deviation = forest_fires['DC'].std()

print(f"DC range: {dc_range}")
print(f"DC mean: {dc_mean}")
print(f"DC standard deviation: {dc_standard_deviation}")

# ISI
isi_range = forest_fires['ISI'].max() - forest_fires['ISI'].min()
isi_mean = forest_fires['ISI'].mean()
isi_standard_deviation = forest_fires['ISI'].std()

print(f"ISI range: {isi_range}")
print(f"ISI mean: {isi_mean}")
print(f"ISI standard deviation: {isi_standard_deviation}")

# temp
temp_range = forest_fires['temp'].max() - forest_fires['temp'].min()
temp_mean = forest_fires['temp'].mean()
temp_standard_deviation = forest_fires['temp'].std()

print(f"temp range: {temp_range}")
print(f"temp mean: {temp_mean}")
print(f"temp standard deviation: {temp_standard_deviation}")

# RH
rh_range = forest_fires['RH'].max() - forest_fires['RH'].min()
rh_mean = forest_fires['RH'].mean()
rh_standard_deviation = forest_fires['RH'].std()

print(f"RH range: {rh_range}")
print(f"RH mean: {rh_mean}")
print(f"RH standard deviation: {rh_standard_deviation}")

# wind
wind_range = forest_fires['wind'].max() - forest_fires['wind'].min()
wind_mean = forest_fires['wind'].mean()
wind_standard_deviation = forest_fires['wind'].std()

print(f"wind range: {wind_range}")
print(f"wind mean: {wind_mean}")
print(f"wind standard deviation: {wind_standard_deviation}")

# rain
rain_range = forest_fires['rain'].max() - forest_fires['rain'].min()
rain_mean = forest_fires['rain'].mean()
rain_standard_deviation = forest_fires['rain'].std()

print(f"rain range: {rain_range}")
print(f"rain mean: {rain_mean}")
print(f"rain standard deviation: {rain_standard_deviation}")

# area
area_range = forest_fires['area'].max() - forest_fires['area'].min()
area_mean = forest_fires['area'].mean()
area_standard_deviation = forest_fires['area'].std()

print(f"area range: {area_range}")
print(f"area mean: {area_mean}")
print(f"area standard deviation: {area_standard_deviation}")

# Find which month have highest number of fires
print(f"-----------------------------------")
fires_count = forest_fires.groupby('month')['area'].sum().idxmax()
print(f"{fires_count} has the highest number of fires.")
```

**(c, 8 points)** Produce boxplots of relative humidity (RH) by month. Your figure will have
a boxplot for every month. Which month has the highest median RH value?

```python
import pandas as pd
import matplotlib.pyplot as plt

# First, show months
month_sorted = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
forest_fires['month'] = pd.Categorical(forest_fires['month'], categories=month_sorted)

# Produce boxplots
forest_fires.boxplot(column = 'RH', by = 'month')
plt.title('Boxplots of relative humidity (RH) by month')
plt.suptitle('')
plt.xlabel('Month')
plt.ylabel('RH')
plt.figure(figsize=(5,5))

# Appropriate picture
plt.show()

# Count and print
print(f"-----------------------------------")
highest_rh = forest_fires.groupby('month')['RH'].mean().idxmax()
print(f"{highest_rh} has the highest median RH value.")
```

**(d, 10 points)** Produce a bar plot to show the count of forest fires in each month for
which wind is greater than 4.9. During which months are high wind forest fires most common?
(Hint: filter data by wind, group data by month and calculate count.)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data by wind
# If area not 0 means that there are fires in this month
wind_filter = forest_fires[(forest_fires['wind'] > 4.9) & (forest_fires['area'] > 0)]
# Group data by month
month_sorted = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
forest_fires['month'] = pd.Categorical(forest_fires['month'], categories=month_sorted)

plt.title('Relationship between wind and fires')
sns.countplot(data = wind_filter, x = 'month', color='blue')
plt.xlabel('Month')
plt.ylabel('Number of forest fires')
plt.figure(figsize=(5,5))

# Appropriate picture
plt.show()

# Count and print
print(f"-----------------------------------")
highest_wind = wind_filter['month'].value_counts().idxmax()
print(f"{highest_wind} are high wind forest fires most common.")
```

**(e, 10 points)** Using the full data set, investigate the predictors graphically, using

relevant variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Using correlation scores
plt.figure(figsize=(5,5))
corr_matrix = forest_fires[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']].corr()['area'][:-1]
corr_matrix.sort_values().plot(marker='o', linestyle='-', color='blue')
plt.title('correlation matrix for forest fires')
plt.xlabel('quantitative')
plt.ylabel('Area of forest fires')
plt.show()
```
