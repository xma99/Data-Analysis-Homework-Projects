**(d, 10 points)** Create a new qualitative variable, called ALevel, by binning the alcohol
variable into two categories (High and Medium). Specifically, divide the data into two groups
based on whether the alcohol level exceeds 11 or not (alcohol greater than 11 is considered High
otherwise it is considered Medium).
Now produce side-by-side boxplots of the ratio of sulphates to chlorides (hint: create a new
variable that calculates sulphates / chlorides) for each of the two ALevel categories. There
should be two boxes on your figure, one for High and one for Medium. How many samples are
in the High category?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Variable into two categories: High and Medium
redwine['ALevel'] = np.where(redwine['alcohol'] > 11, 'High', 'Medium')
# Count each class
count_High = (redwine['ALevel'] == 'High').sum()
count_Medium = (redwine['ALevel'] == 'Medium').sum()
print(f"In ALevel, High have: {count_High}")
print(f"In ALevel, Medium have: {count_Medium}")
# Plot
plt.title('Wine ALevel: Ratio of sulphates to chlorides')
redwine['sulphates_chlorides'] = redwine['sulphates'] / redwine['chlorides']
sns.boxplot(x = 'ALevel', y = 'sulphates_chlorides', data = redwine)
plt.figure(figsize=(5,5))
plt.show()
```

**(e, 8 points)** Produce a histogram showing the fixed_acidity numbers for both High and
Medium (ALevel) wine samples. You may choose to show both on a single plot (using side by

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Variable into two categories: High and Medium
redwine['ALevel'] = np.where(redwine['alcohol'] > 11, 'High', 'Medium')
# Plot histogram showing the fixed_acidity numbers for both High and Medium (ALevel)
plt.title('Wine fixed_acidity numbers for High(ALevel)')
sns.histplot(data=redwine[redwine['ALevel'] == 'High'], x='fixed_acidity')
plt.ylabel('High_Count')
plt.figure(figsize=(5,5))

plt.title('Wine fixed_acidity numbers for Medium(ALevel)')
sns.histplot(data=redwine[redwine['ALevel'] == 'Medium'], x='fixed_acidity')
plt.ylabel('Medium_Count')
plt.figure(figsize=(5,5))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Draw a pie chart to categorize the quality of wine samples
# Since the mean of the quality is known to be 5.6 in the previous calculations
# Set the quality below 5 as failing, greater than or equal to 5 and less than or equal to 6 as passing, and greater than 6 as good

def classify(wine_quality):
  if wine_quality < 5:
    return 'Fail'
  elif 5 <= wine_quality <=6:
    return 'Pass'
  else:
    return 'Good'

# Draw pie
redwine['result'] = redwine['quality'].apply(classify)
quality_pie = redwine['result'].value_counts()

plt.figure(figsize=(5,5))
quality_pie.plot.pie(autopct='%1.1f%%', startangle=90, colors=['green', 'yellow', 'red'])
plt.title('Categorize the quality of wine samples')
plt.show()
```

**Answer**

It's still possible to see by pH that people do not like acidic wines, even though line graphs don't seem appropriate for this data. The pie chart shows that using the average as a passing standard is a good choice, since it looks like the passing rate is normal.
